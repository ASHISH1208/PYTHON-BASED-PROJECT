```python
#Group the June-September rainfall as South-West Monsoon and October-December as North-East Monsoon  

South_West_M=list(dataframe['Jun-Sep'])
dataframe['South-West Monsoon']=South_West_M
North_East_M=list(dataframe['Oct-Dec'])
dataframe['North-East Monsoon']=North_East_M
dataframe
```

```python
newdf=dataframe.groupby(['REGION','YEAR']).sum()
ndf=newdf[['South-West Monsoon','North-East Monsoon']]
ndf.to_csv('TypeMonsson.csv')
ndf
```

```python
ndf['S-W prev1'] = ndf['South-West Monsoon']
ndf['S-W prev1'] = ndf['S-W prev1'].shift(1)
ndf['S-W prev2'] = ndf['South-West Monsoon']
ndf['S-W prev2'] = ndf['S-W prev2'].shift(2)
ndf['S-W prev3'] = ndf['South-West Monsoon']
ndf['S-W prev3'] = ndf['S-W prev3'].shift(3)
ndf['N-E prev1'] = ndf['North-East Monsoon']
ndf['N-E prev1'] = ndf['N-E prev1'].shift(1)
ndf['N-E prev2'] = ndf['North-East Monsoon']
ndf['N-E prev2'] = ndf['N-E prev1'].shift(2)
ndf['N-E prev3'] = ndf['North-East Monsoon']
ndf['N-E prev3'] = ndf['N-E prev1'].shift(3)
ndf.fillna(0,inplace=True)
ndf.to_csv('Forcastprev.csv')
ndf
```
